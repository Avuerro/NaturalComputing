### Very simple PSO algorithm

```python
def simple_PSO(x_init, v_init, n_iterations, omega, alpha_1, alpha_2, r_1, r_2):
    # Initialize position x and velocity v
    x = x_init
    v = v_init
    
    # Initialize the individual/global best and a list that keeps track of all values of x
    x_best = x
    x_list = [x]
    
    # Loop over iterations
    for i in range(n_iterations):
        # Update v and x (x_best is both the individual and global best, so we can simplify the update rule)
        v = omega*v + (alpha_1*r_1 + alpha_2*r_2)*(x_best-x)
        x = x+v
        
        # Append new x to the list of x values
        x_list.append(x)
        
        # If f(x) = x^2 is better (lower) than f(x_best) = x_best^2, update x_best
        if x**2 < x_best**2:
            x_best = x
    
    # return the best position found and the list of position values
    return x_best, x_list
```

### Running the algorithm

```python
x_init = 20
v_init = 10
n_iterations = 100
```

```python
# Run an example where the result is suboptimal, and an example where the result is optimal
x_best_suboptimal, x_list_suboptimal = simple_PSO(x_init, v_init, n_iterations, 0.5, 1.5, 1.5, 0.5, 0.5)
x_best_optimal, x_list_optimal = simple_PSO(x_init, v_init, n_iterations, 0.7, 1.5, 1.5, 1, 1)
```

### Plotting results

```python
def plot_results(x_list, x_best):
    plt.plot(x_list)
    plt.xlabel("Number of iterations")
    plt.ylabel("Position of the particle")
    plt.title(f"Trajectory of a single particle minimizing f(x) = x^2\n"\
              f"{'Optimal' if x_best < 0.000001 else 'Suboptimal'} case: final best value of x is {x_best:.3f}")
    plt.show()
```

```python
# Plot results for suboptimal and optimal example
plot_results(x_list_suboptimal, x_best_suboptimal)
plot_results(x_list_optimal, x_best_optimal)
```
