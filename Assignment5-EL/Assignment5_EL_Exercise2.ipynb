### Imports

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

### Jury theorem probability function

```python
def P_cmv(n,p):
    total_p = 0
    for i in range(math.floor(n/2) + 1):
        total_p += p**(n-i) * (1-p)**i * math.comb(n,i)
    return total_p
```

### Exercise 2b: probability of a correct majority vote (cmv) for 31 students with p = 0.6

```python
print(P_cmv(31,0.6))
```

### Exercise 2c: graph of probability of cmv as a function of p and the number of individuals

```python
def plotP_cmv(max_n, step, ps):
    # Plot line for every probability
    for p in ps:
        p_correct_values = [P_cmv(n,p) for n in range(1,max_n+1,step)]
        plt.plot(range(1,max_n+1,step), p_correct_values, label = f"p = {p}")

    # Set x and y labels, and add legend
    plt.xlabel("n")
    plt.ylabel("P_cmv")
    plt.title(f"P_cmv as a function of number of individuals {'(only odd values of n)' if step==2 else '(odd and even values of n)'}")
    if len(ps) > 1:
        plt.legend(loc="upper left", bbox_to_anchor=(1, 0.85), )

    # Plot the result
    plt.show()
```

```python
# Plot for all probabilities between 0.55 and 0.95 (with a step of 0.5); only odd values of n
ps = [0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]
plotP_cmv(100,2,ps)
```

```python
# Plot for p=0.6; both odd and even values of n
plotP_cmv(100,1,[0.6])
```
