### Jury theorem probability function

```python
def P_cmv(n,p):
    # Compute P_cvm
    p_cmv = 0
    for i in range(math.floor(n/2) + 1):
        p_cmv += p**(n-i) * (1-p)**i * math.comb(n,i)
    
    # Correct for even values of n
    if not n%2:
        p_cmv -= 0.5 * p**(n//2) * (1-p)**(n//2) * math.comb(n,n//2) 
    
    # Return the calculated value
    return p_cmv
```

### Exercise 2b: probability of a correct majority vote (cmv) for 31 students with p = 0.6

```python
print(P_cmv(31,0.6))
```

### Exercise 2c: graph of probability of cmv as a function of p and the number of individuals

```python
def plotP_cmv(max_n):
    # Set size of the plot
    plt.figure(figsize=(8,5))
    
    # Plot line for every probability between 0.55 and 0.95 (with a step of 0.5)
    for p in [0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95]:
        p_cmv_values = [P_cmv(n,p) for n in range(1,max_n+1)]
        plt.plot(range(1,max_n+1), p_cmv_values, label = f"p = {p}")

    # Prettify the plot
    plt.xlabel("n")
    plt.ylabel("P_cmv")
    plt.title(f"P_cmv as a function of number of individuals")
    plt.legend(loc="upper left", bbox_to_anchor=(1, 0.85))
    plt.grid()

    # Show the result
    plt.show()
```

```python
plotP_cmv(100)
```
